```python
import matplotlib.pyplot as plt
import numpy as np

species = ("FPR", "TPR", "AUC")
penguin_means = {
    'w/o T on AIG': (2.82, 85.18, 98.21),
    'IRLM on AIG': (0.39, 92.98, 98.71),
    'w/o T on ACG': (5.05, 79.07, 96.90),
    'IRLM on ACG': (2.65, 95.35, 99.42),
}

x = np.arange(len(species))  # the label locations
width = 0.22  # the width of the bars
multiplier = -0.5

fig, ax = plt.subplots(layout='constrained',figsize = (7,4))

colors = ['#B2A4FF','#F7C8E0','#B6FFCE','#7FB7E0']
i=0
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, color = colors[i], label=attribute)
#     ax.bar_label(rects, padding=4,size = 12)
    i=i+1
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('values (%)',fontsize = 18)
ax.set_title('')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncol=2)
ax.set_ylim(0, 125)
ax.legend(fontsize=14.5)
ax.tick_params(axis='both', which='major', labelsize=16)
plt.savefig("ablation_time.png",dpi=500,bbox_inches = 'tight')
plt.show()
```

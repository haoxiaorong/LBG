```python
import matplotlib.pyplot as plt
from pylab import *                                 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
plt.subplots(figsize=(6.4,5.2))
x= np.arange(6,26,2)
yFPR=np.array([26.12,13.54,8.9,6.55,4.96,3.92,3.31,2.81,2.34,1.89])
yTPR=np.array([98.91,98.91,98.75,98.29,97.51,95.48,93.46,92.52,91.28,87.85])
A, = plt.plot(x, yFPR, marker='.', mec='blue', mfc='w',label=u'FPR',linewidth =2,markersize=10)
B, = plt.plot(x, yTPR, marker='*', mec='orange', mfc='w',label=u'TPR',linewidth = 2,markersize=10)
font1 = {'family':'Times New Roman',
         'weight':'normal',
         'size':24
         }
legend = plt.legend(handles=[A,B], prop=font1,loc='lower center',ncol=1,bbox_to_anchor=(0.8,0.1))
x_major_locator=MultipleLocator(1)
y_major_locator=MultipleLocator(20)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
plt.margins(0)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.subplots_adjust(bottom=0.18,left = 0.14)
plt.xlabel("radius",fontsize = 32) 
plt.ylabel("scores(%)",fontsize = 32)
plt.ylim(0,105)
plt.xlim(5,25)
x_major_locator=MultipleLocator(5)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.tick_params(axis='both', which='major', labelsize=26)
plt.gca().yaxis.get_offset_text().set_fontsize(24)
plt.title('',fontsize = 12)
plt.savefig("radius_change.png",dpi=500,bbox_inches = 'tight')
plt.show()
```

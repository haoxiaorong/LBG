```python
import matplotlib.pyplot as plt
from pylab import *                                 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
plt.subplots(figsize=(6.4,5.2))
x= np.arange(10,110,10)
yFPR=np.array([0.02650721,0.0372114,0.071860756,0.0959417,0.0979734,0.04154886,0.06151053,0.074406059,0.0428469,0.02263117])*100
yTPR=np.array([0.95348837,1,0.860465,0.8139534,0.8372093,0.74418604,0.69767441,0.81395348,0.697674418,0.8139534])*100
yAUC=np.array([0.99419652,0.99197025,0.9611305,0.952488,0.94158278,0.9609229,0.91856873,0.94565202,0.92104114,0.96906279])*100
A, = plt.plot(x, yFPR, marker='.', mec='blue', mfc='w',label=u'FPR',linewidth = 2,markersize=15)
B, = plt.plot(x, yTPR, marker='*', mec='orange', mfc='w',label=u'TPR',linewidth = 2,markersize=10)
C, = plt.plot(x, yAUC, marker='^', mec='green', mfc='w',label=u'AUC',linewidth = 2,markersize=10)
font1 = {'family':'Times New Roman',
         'weight':'normal',
         'size':26
         }
legend = plt.legend(handles=[A,B,C], prop=font1,loc='lower center',ncol=1,bbox_to_anchor=(0.78,0.1))
x_major_locator=MultipleLocator(20)
y_major_locator=MultipleLocator(20)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
plt.margins(0)
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)
plt.subplots_adjust(bottom=0.18,left = 0.14)
plt.xlabel("rank by node occurrence",fontsize = 32) 
plt.ylabel("scores(%)",fontsize = 32)
plt.ylim(0,105)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
#plt.title("\") 
#plt.axis('off')
plt.title('',fontsize = 12)
plt.savefig("anom_node_rank_score_lanl.png",dpi=500,bbox_inches = 'tight')
plt.show()
```
